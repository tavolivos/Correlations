```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from pylab import rcParams
rl=linear_model.LinearRegression()
```

```python
data = pd.read_csv("data_11.csv")
data
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 7, 7

x = data["BE1"]
y = data["Eff"]

plt.xlabel('BE')
plt.ylabel('Kcat')
plt.scatter(x,y, color='red', marker='o')

cc = round(x.corr(y), 3)
print ("Coeficiente de correlacion:", cc)

X = x[:, np.newaxis]
y = data["Kcat"]

rl.fit(X,y)

plt.plot(X,rl.predict(X), color='blue')
plt.show()
```

```python
# Shapiro-Wilk Test
# normality test
from scipy.stats import shapiro

data = df['Act']

plt.hist(data)

stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```
